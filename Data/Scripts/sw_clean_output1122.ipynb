```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1= pd.read_csv('../Raw/output/USPRIV.csv')
df1['DATE'] = pd.to_datetime(df1['DATE'])
df1 = df1[df1['DATE'] >= '1960-01-01']
df1.rename(columns = {'USPRIV_PCH' : 'Total_private_pch', 'DATE': 'Month'}, inplace = True)
print(df1.shape)
# df.to_csv('../Employ_total_private_1939.csv')

plt.plot(df1['Month'], df1['Total_private_pch'])
plt.title('Monthly % change in total private employment, seasonally adjusted')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.show()

print(df1.head(10))
```

```python
df2= pd.read_csv('../Raw/output/USCONS.csv')
df2['DATE'] = pd.to_datetime(df2['DATE'])
df2 = df2[df2['DATE'] >= '1960-01-01']
df2.rename(columns = {'USCONS_PCH' : 'Construction_pch', 'DATE': 'Month'}, inplace = True)
print(df2.shape)
# df.to_csv('../Employ_total_private_1939.csv')

plt.plot(df2['Month'], df2['Construction_pch'])
plt.title('Monthly % change in construction employment, seasonally adjusted')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.show()
```

```python
df3= pd.read_csv('../Raw/output/UNRATE.csv')
# df3
df3['DATE'] = pd.to_datetime(df3['DATE'])
df3 = df3[df3['DATE'] >= '1960-01-01']
df3.rename(columns = {'UNRATE' : 'Unemploy_rate', 'DATE': 'Month'}, inplace = True)
print(df3.shape)
# df.to_csv('../Employ_total_private_1939.csv')

plt.plot(df3['Month'], df3['Unemploy_rate'])
plt.title('Monthly unemployment rate, seasonally adjusted')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.show()
```

```python
df = df1.merge(df2, how = 'left', on = 'Month')
df = df.merge(df3, how = 'left', on = 'Month')
df.isnull().any().any()
df.to_csv('../Output_employpch_unemployrate_1960.csv')
df.shape
df
```
