```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_input = pd.read_csv('../Input_features_1960_no_covid.csv')
df_input['Month'] = pd.to_datetime(df_input['Month'])
df_input
```

```python
df_output = pd.read_csv('../Employment_pch_1960_no_covid.csv')
df_output['Month'] = pd.to_datetime(df_output['Month'])
```

```python
plt.plot(df_output['Month'], df_output['Total_private_pch'])
plt.xlabel('Time (Year)')
plt.ylabel('Eployment percentage change (%)')
```

```python
plt.plot(df_input['Month'], df_input['Loan_pch'], label = 'Consumer Loan', linestyle = ':', color = 'gray')
plt.plot(df_input['Month'], df_input['GDP_pch'], label = 'GDP', color = 'green')
plt.plot(df_input['Month'], df_input['CPI_pc1'], label = 'CPI', color = 'orange')
plt.legend()
plt.xlabel('Time (Year)')
plt.ylabel('Percentage change (%)')
```
