```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Total private
df = pd.read_csv('../Raw/output/USTotalPrivate.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
# df = df[(df['DATE'] < '2020-04-01') | (df['DATE'] > '2020-08-01')  ]
df_diff = df.diff(axis = 0)
df_diff['DATE'] = df['DATE']
df_diff.rename(columns = {'USPRIV' : 'Total_private', 'DATE': 'Month'}, inplace = True)
df1 = df_diff.iloc[1:,:]
print(df1.shape)
# df.to_csv('../Employ_total_private_1939.csv')

plt.plot(df1['Month'], df1['Total_private'])
plt.title('Monthly change in total private employment, seasonally adjusted')
plt.xlabel('Year')
plt.ylabel('Thousands of persons')
plt.show()
```

```python
## information sector
df = pd.read_csv('../Raw/output/USInfo.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
# df = df[(df['DATE'] < '2020-04-01') | (df['DATE'] > '2020-08-01')  ]
df_diff = df.diff(axis = 0)
df_diff['DATE'] = df['DATE']
df_diff.rename(columns = {'USINFO' : 'Information', 'DATE': 'Month'}, inplace = True)
df2 = df_diff.iloc[1:,:]
print(df2.shape)
# df.to_csv('../Employ_information_1939.csv')

plt.plot(df2['Month'], df2['Information'])
plt.title('Monthly change in private employment, seasonally adjusted')
plt.xlabel('Year')
plt.ylabel('Thousands of persons')
plt.show()
```
